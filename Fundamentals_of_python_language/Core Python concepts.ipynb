```python
#  B10. Write a Python program to get the Factorial number of given number

num = int(input("Enter a number: "))    
factorial = 1    
if num < 0:    
   print(" Factorial does not exist for negative numbers")    
elif num == 0:    
   print("The factorial of 0 is 1")    
else:    
   for i in range(1,num + 1):    
       factorial = factorial*i    
   print("The factorial of",num,"is",factorial)  
```

```python
# B11. Write a Python program to get the Fibonacci series of given range

n_terms = int(input ("How many terms the user wants to print? "))  
  
  
n_1 = 0  
n_2 = 1  
count = 0  
  
  
if n_terms <= 0:  
    print ("Please enter a positive integer, the given number is not valid")  

elif n_terms == 1:  
    print ("The Fibonacci sequence of the numbers up to", n_terms, ": ")  
    print(n_1)  

else:  
    print ("The fibonacci sequence of the numbers is:")  
    while count < n_terms:  
        print(n_1)  
        nth = n_1 + n_2
        n_1 = n_2  
        n_2 = nth  
        count += 1
```

```python
# Assignment Level Advance

# A1. Write a Python program to sort three integers without using conditional statements and loops

a = int(input("Input the first number: "))
y = int(input("Input the second number: "))
z = int(input("Input the third number: "))

a1 = min(a,y,z)
a3 = max(a,y,z)

a2 = (a+y+z) - a1 - a3
print("Number in sorted order: ", a1,a2,a3)
```
