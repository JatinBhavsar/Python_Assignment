```python
# B1. Write a python functions to calculate the factorial of a number (a non-negative integer)

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
n=int(input("Please Enter a factiorial Number : "))
print(factorial(n))
```

```python
# B4.  Write a Python function that checks whether a passed string is palindrome or not 

def isPalindrome(string):
	left_pos = 0
	right_pos = len(string) - 1
	
	while right_pos >= left_pos:
		if not string[left_pos] == string[right_pos]:
			return False
		left_pos += 1
		right_pos -= 1
	return True
print(isPalindrome('aza'))
```
